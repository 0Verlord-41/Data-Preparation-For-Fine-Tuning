```python
import pandas as pd
```

```python
new_path = "/content/fighter_data.csv"

df_scrapped = pd.read_csv(new_path)
df_scrapped.head(6)
```

```python
df = df_scrapped.drop_duplicates(subset=['name','profile_url'])
df.head(6)
```

```python
import numpy as np
import re

# Height: convert to inches
def to_inches(height_str):
  try:
    feet, inches = re.match(r"(\d+)'\s*(\d+)", height_str).groups()
    return (int(feet) * 12) + int(inches)
  except:
    return np.nan

df['height_inches'] = df['height'].apply(to_inches)


# Reach: convert to float and replace '--' with NaN
df['reach'] = df["reach"].replace("--", np.nan)

def format_reach(reach_str):
  try:
    return float(reach_str.replace('"', ''))
  except:
    return np.nan

df['reach'] = df['reach'].apply(format_reach)
df.head(6)
```

```python

# Clean percentages
df["Str. Acc."] = df["Str. Acc."].str.replace('%', '').astype(float) / 100
df["Str. Def"] = df["Str. Def"].str.replace('%', '').astype(float) / 100

df.head(6)
```

```python
# draws: clean the draws column
df['draws'] = df['draws'].apply(lambda x: int(re.search(r"\d+", str(x)).group()))

df.head(6)
```

```python
#create total fights and win rate
df['total_fights'] = df['wins'] + df['losses'] + df['draws']
df['win_rate'] = df['wins'] / df['total_fights']

df.head(6)
```

```python
# Drop NaNs or impute
df = df.dropna(subset=["height_inches", "reach", "Str. Acc.", "Str. Def"])
df.head(6)
```

```python
#dropping original height column

df.drop(columns=['height'], inplace=True)

df.head(6)
```

```python
df.describe()
```

```python
df.to_csv("final_fighter_data.csv", index=False)
```

## Matchups creation


```python
import pandas as pd
import random
from itertools import combinations

latest_path = "/content/final_fighter_data.csv"

df_final = pd.read_csv(latest_path)
df_final.set_index('name', inplace=True)

df_final.head(6)
```

```python
fighters = df_final.index.tolist()
fighter_pairs = list(combinations(fighters, 2))

print(fighter_pairs[:10])
```

```python
sampled_matchups = random.sample(fighter_pairs, 1000)
```

```python
import numpy as np
rows = []

for fighter_a, fighter_b in sampled_matchups:
  a = df_final.loc[fighter_a]
  b = df_final.loc[fighter_b]

  input_text = (
      f"Fighter A: {fighter_a} | Height: {a['height_inches']} in | Reach: {a['reach']} in | Str. Acc: {a['Str. Acc.']:.2f} | Str. Def: {a['Str. Def']:.2f} | SLpm: {a['SLpM']} | SApM: {a['SApM']} || "
      f"Fighter B: {fighter_b} | Height: {b['height_inches']} in | Reach: {b['reach']} in | Str. Acc: {b['Str. Acc.']:.2f} | Str. Def: {b['Str. Def']:.2f} | SLpm: {b['SLpM']} | SApM: {b['SApM']}"
  )

  label = 0 if np.round(a['win_rate'], decimals=3)>np.round(b['win_rate'], decimals=3) else 1

  rows.append({
      'fighter_a': fighter_a,
      'fighter_b': fighter_b,
      'input_text': input_text,
      'label': label
  })

  df_matchups = pd.DataFrame(rows)
```

```python
df_matchups.info()

df_matchups.head(6)
```

```python
df_matchups.to_csv("final_matchups.csv", index=False)
```
