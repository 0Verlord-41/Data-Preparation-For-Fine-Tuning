```python
import pandas as pd

file_path = "/root/.cache/kagglehub/datasets/ustice/ufc-records-dataset/versions/5/UFC_Records.csv"
df_bonus = pd.read_csv(file_path, index_col=0)
```

```python
df_bonus.head()
```

```python
df_bonus.rename(columns={df_bonus.columns[0]: 'Type of Record'}, inplace=True)

df_bonus.head()
```

```python
df_bonus.info()
```

```python
# Converting df to a wide format from long format
fighter_stats = df_bonus.pivot(index="Name", columns="Type of Record", values="Total").reset_index()

fighter_stats.head()
```

```python
fighter_stats.columns.name = None
fighter_stats = fighter_stats.rename(columns={
    "Name": "Fighter",
    "Total Fights": "total_fights",
    "Wins": "wins",
    "Losses": "losses",
    "KOs": "kos",
})

fighter_stats.head()
```

```python
fighter_stats[fighter_stats["Fighter"]=="Jim Miller"]
```

```python
from itertools import combinations
import random

# Generate matchups (random or from real fights)
fighters = fighter_stats["Fighter"].tolist()
matchups = list(combinations(fighters, 2))  # every pair of fighters

# Sample for faster testing
matchups = random.sample(matchups, 500)  # reduce for small training set
```
